```python
from IPython.display import clear_output
def start():
    clear_output()
    print("Welcome to Tic-Tac-Toe")
    player_1 = input("Enter player 1 name\n")
    player_2 = input("Enter player 2 name\n")

    token = input(f"{player_1} choose your token\n")
    if token == "x":
        print(f"{player_1} will start")
        global start_p 
        start_p = f"{player_1}"
        global start_p2 
        start_p2 = f"{player_2}"
    else:
        print(f"{player_2} will start") 
        start_p = f"{player_2}"
        start_p2 = f"{player_1}"

global place_x
place_x = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']        
        
def display_board():    
    print( "       |       |   ")
    print(  place_x[1] +'      |   '+   place_x[2]   +'   |  '+   place_x[3])
    print( "       |       |   ")
    print("-----------------------")
    print( "       |       |   ")
    print(  place_x[4] +'      |   '+   place_x[5]   +'   |   '+   place_x[6])
    print( "       |       |   ")
    print( "-----------------------")
    print( "       |       |   ")
    print(  place_x[7] +'      |   '+   place_x[8]   +'   |   '+   place_x[9])
    print( "       |       |   ")
    
flag = False
def check(): 
    global flag
    if place_x[1:4]==list('X'*3) or place_x[4:7]==list('X'*3) or place_x[7:]==list('X'*3):
        print(start_p + " won the game")
        flag=True
    elif place_x[1:4]==list('O'*3) or place_x[4:7]==list('O'*3) or place_x[7:]==list('O'*3):
        print(start_p2 + " won the game")
        flag=True
    else:
        for i in range(1,4):
            if i==1:
                if place_x[i]==place_x[i+4]==place_x[i+8]=='X':
                    print(start_p + " won the game")
                    flag=True
                    break
                elif place_x[i]==place_x[i+4]==place_x[i+8]=='O':
                    print(start_p2 + " won the game")
                    flag=True
                    break
                elif place_x[i]==place_x[i+3]==place_x[i+6]=='X':
                    print(start_p + " won the game")
                    flag=True
                    break
                elif place_x[i]==place_x[i+3]==place_x[i+6]=='O':
                    print(start_p2 + " won the game")
                    flag=True
                    break
            if i==2:
                if place_x[i]==place_x[i+3]==place_x[i+6]=='X':
                    print(start_p + " won the game")
                    flag=True
                    break
                elif place_x[i]==place_x[i+3]==place_x[i+6]=='O':
                    print(start_p2 + " won the game")
                    flag=True
                    break
                else:
                    continue
            if i==3:
                if place_x[i]==place_x[i+2]==place_x[i+4]=='X':
                    print(start_p + " won the game")
                    flag=True
                    break
                elif place_x[i]==place_x[i+2]==place_x[i+4]=='O':
                    print(start_p2 + " won the game")
                    flag=True
                    break
                elif place_x[i]==place_x[i+3]==place_x[i+6]=='X':
                    print(start_p + " won the game")
                    flag=True
                    break
                elif place_x[i]==place_x[i+3]==place_x[i+6]=='O':
                    print(start_p2 + " won the game")
                    flag=True
                    break
                    
def play():
    x = 0
    while x <= 8:
        place_choosed = input(start_p + ": Choose your position  (1,9)\n")
        place_x[int(place_choosed)] = 'X'
        clear_output()
        display_board()
        x+=1
    
        if x>=5:
            check()
        if flag:
            break
        if x==9:
            if not flag:
                print("DRAW!")
                break
            break
        
        place_choosed = input(start_p2 + ": Choose next position  (1,9)\n")
        place_x[int(place_choosed)] = 'O'
        clear_output()
        display_board()
        x+=1
    
        if x>=5:
            check()
        if flag:
            break
            
def replay():
    reply=input("Do you want to PLAY AGAIN?\n")
    if reply=="yes":
        start()
        global place_x
        place_x = ['#','','','','','','','','','']   
        display_board()
        check()
        play()
        replay()
    else:
        print("Thank You for Playing")
```

```python
start()
display_board()
check()
play()
replay()
```
